```python
import numpy as np
```

```python
points=np.arange(-5,5,0.01)
```

```python
points
```

```python
x,y=np.meshgrid(points,points) # it give two 1D array and produce 2D matrix
```

```python
y
```

```python
z=np.sqrt(x**2,y**2)
```

```python
z
```

```python
# use matplotlib to visulization of given function
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(z,cmap=plt.cm.gray); plt.colorbar()
```

```python
plt.title("Image plot of $\sqrt{x^2+y^2}$ for a grid of values")
```

```python
# expessing conditional logic as array Operation
```

```python
x2=np.arange(1.1,1.6,0.1)
```

```python
x2
```

```python
y2=np.array([2.1,2.2,2.3,2.4,2.5])
```

```python
y2
```

```python
cond2=np.array([True, False, True, True, False])
```

```python
result=[(x if c else y) for x,y,c in zip(x2,y2,cond2)]
```

```python
result
```

```python
result2=np.where(cond2,x2,y2)
```

```python
result2
```
