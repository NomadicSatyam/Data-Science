```python
import numpy as np
x=np.arange(10)
y=x*2
z=x**2

import matplotlib.pyplot as plt
%matplotlib inline
plt.show()
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')
```

```python
fig=plt.figure()

ax1=fig.add_axes([0,0,1,1])
ax2= fig.add_axes([0.2,0.5,0.2,0.2])
ax1.plot(x,y)
ax2.plot(x,y)
```

```python
fig=plt.figure()
ax1=fig.add_axes([0,0,1,1])
ax2=fig.add_axes([0.2,.5,.4,.4])

#Large
ax.plot(x,z)
ax.set_xlabel('x')
ax.set_ylabel('z')

#Insert
ax.plot(x,y)
ax2.set_title('zoom')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_xlim([20,22])
ax2.set_xlim([30,50])
```

```python
fig,axes=plt.subplots(1,2,figsize=(12,2))
axes[0].plot(x,y,color='blue',lw=3,ls='-')
axes[1].plot(x,y,color='red',lw=3,ls='--')
```
