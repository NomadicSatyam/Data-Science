```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
import random
```

```python
classes=['python','R','Ml','AI','DS']
```

```python
class1_student=[10,20,30,40,50]
class2_student=[40,50,20,20,10]
class3_student=[35,5,30,10,10]
```

```python
plt.bar(classes,class1_student,width=0.2,align='edge',color='y',edgecolor='m'
        ,linewidth=5,alpha=0.7,linestyle='--',label='Class 1 student',visible=True)
```

```python
plt.barh(classes,class1_student,)
plt.legend()
plt.show()
```

```python
style.use('ggplot')
```

```python
plt.bar(classes,class1_student,width=0.2,align='edge',color='y',label='Class 1 student')
plt.title("Student Barchart")
plt.xlabel('classes')
plt.ylabel('No of student age')
plt.legend()
plt.show()
```

```python
plt.bar(classes,class2_student,width=0.2,align='edge',color='g',label='Class 2 student')
plt.title("Student Barchart")
plt.xlabel('classes')
plt.ylabel('No of student age')
plt.legend()
plt.show()
```

```python
style.use('ggplot')
```

```python
classes_index=np.arange(len(classes))
width=0.2
```

```python
plt.bar(classes_index,class1_student,width,align='edge',color='y',label='Class 1 student')

plt.bar(classes_index+width,class2_student,width,align='edge',color='g',label='Class 2 student')

plt.bar(classes_index+width+width,class3_student,width,align='edge',color='m',label='Class 2 student')

plt.xticks(classes_index+width,classes,rotation=0)

plt.title("Student Barchart")
plt.xlabel('classes')
plt.ylabel('No of student age')
plt.legend()
plt.show()
```
