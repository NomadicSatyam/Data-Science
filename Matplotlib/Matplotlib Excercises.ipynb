```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.show()
```

Data

```python
x=np.arange(10)
y=x*2
z=x**2
```


''' Excercise 1

    Follow along with these step
    
    
'''

     >Create a figure object called fig using plt.figure(),
     
     >use add_axes to add an axis to the figure canvas at,
      [0,0,1,1] call this new axis ax, 
      
     >plot (x,y) on that axes and set the labels and titles
      to match the plot below 

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')
```

''' Excercise 2

   > Create a figure object and put two axes on it ,
   
     ax1 and ax2 located at [0,0,1,1] and [0.2,0.5,.2,.2]
'''

```python
fig=plt.figure()

ax1=fig.add_axes([0,0,1,1])
ax2= fig.add_axes([0.2,0.5,0.2,0.2])
ax1.plot(x,y,color='red')
ax2.plot(x,y)
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax2=fig.add_axes([0.2,.5,.4,.4])

#Large
ax.plot(x,z)
ax.set_xlabel('x')
ax.set_ylabel('z')

#Insert
ax.plot(x,y)
ax2.set_title('zoom')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_xlim([20,22])
ax2.set_xlim([30,50])
```
